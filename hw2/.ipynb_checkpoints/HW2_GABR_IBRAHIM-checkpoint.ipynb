```python
def fill_values(df):
    male = np.nanmean(np.asarray(df["male"]))
    head_earnings = np.nanmean(np.asarray(df['head_earnings']))
    bmi_40 = np.nanmean(np.asarray(df["bmi_40"]))
    no_family_in_nbhd = np.nanmean(np.asarray(df['no_family_in_nbhd']))
    anxiety = np.nanmean(np.asarray(df["anxiety"]))
    return np.asarray([male, head_earnings,bmi_40, no_family_in_nbhd, anxiety]) 
```

```python
def generate_stats(df, charac_lst):
    stat_dict = {}
    for c in charac_lst:
        mu = np.nanmean(np.asarray(df[c])) # nanmean to ensure NaN values are omitted
        s = np.nanstd(np.asarray(df[c])) # nanstd to ensure NaN values are omitted
        N = df[c].notnull().sum() # calculates number of non-NaN values
        stat_dict[c] = (mu,s,N)
    return stat_dict
```

```python
def p_values(df1, df2, charac_lst):
    p_dict = {}
    group_1_stats = generate_stats(df1, charac_lst)
    group_2_stats = generate_stats(df2, charac_lst)
    
    for c in charac_lst:
        m1, s1, n1 = group_1_stats[c]
        m2, s2, n2 = group_2_stats[c]
        p_dict[c] = ttest(m1, s1, n1, m2, s2, n2).pvalue
    return p_dict
```

```python
E = fill_values(experimental_df)
S = fill_values(section8_df)
C = fill_values(control_df)
E_C_diff = E - C
S_C_diff = S - C
S_C_p_vals = p_values(section8_df, control_df, col_list)
E_C_p_vals = p_values(experimental_df, control_df, col_list)
```

```python
table = pd.DataFrame([dict(zip(col_list,E)), dict(zip(col_list,S)),
             dict(zip(col_list,C)), dict(zip(col_list,E_C_diff)),
             E_C_p_vals, dict(zip(col_list, S_C_diff)),
             S_C_p_vals]).T

table.columns = ["Experimental group (E) mean", "Section 8 group (S) mean",
                                      "Control group (C) mean", "Difference E-C",
                                      "p-value E-C", "Difference S-C","p-value S-C"]
table = table.reindex(index=col_list).round(3)

d = dict(selector="th",
    props=[('text-align', 'center')])

table.style.set_properties(**{'width':'10em', 'text-align':'center'})\
        .set_table_styles([d])
```

## Question 7

All of the coefficients match those that we calculated by comparing the means directly except for `absence_mental_probs_z`.

The std error for every bivariate regression is 0.04. This is small, relative to the magnitude of the coefficients - this is a good sign as large standard errors lead to imprecise coefficients as would be evident by having a large confidence interval.

Significant at 90% level:

- `overall_index_z`

Significant at 95% level:

- `absence_mental_probs_z`

- `psychol_distress_z`

```python
# Experimental Group

lst = ["overall_index_z", "economic_self_sufficiency_z", 
       "absence_phys_probs_z", "absence_mental_probs_z", 
       "psychol_distress_z"]

experimental_df = data[data['group_section8'] == 0.0]

for y in lst:
    indep = experimental_df["group_experimental"].values
    X = sm.add_constant(indep)
    outcome = experimental_df[y].values

    model = sm.OLS(outcome, X, missing='drop')
    results = model.fit()
    coeff = round(results.params[-1],3)
    p_val = round(results.pvalues[-1],3)
    std_error = round(results.bse[-1],3)
    r2 = round(results.rsquared,3)
    print(f"coeff for {y}: ", coeff)
    print(f"Std Error for {y}: ", std_error)
    print(f"P-Value for {y}: ", p_val)
    print(f"R^2 for {y}: ", r2)
    if p_val <= 0.05:
        print(f"{y} is significant at 95% Level")
    elif p_val <= 0.1:
        print(f"{y} is significant at 90% Level")
    else:
        print(f"{y} is not significant at 95% or 90 % Level")
    print()
```

## Question 10

Please see code output below for the table entries.

```python
def encode(array):
    return array.apply(lambda x: 0 if x[0] == "0" else 1).astype('float64')
```

```python
#selecting features and encoding dummies

excluded_features = ["city_NYC", "age_1_35", "no_high_school", "diabetes"]
control_features = [x for x in data.columns[10:] if x not in excluded_features]
control_features.insert(0, "group_experimental")

for feature in control_features:
    if data[feature].dtype == np.dtype('float32'):
        data[feature] = data[feature].astype('float64')
    elif data[feature].dtype != np.dtype('float64'):
        data[feature] = encode(data[feature])
```

```python
# Experimental Group
lst = ["overall_index_z", "economic_self_sufficiency_z",
       "absence_phys_probs_z", "absence_mental_probs_z",
       "psychol_distress_z"]

experimental_df = data[data['group_section8'] == 0.0]

for y in lst:
    indep = experimental_df[control_features].values
    X = sm.add_constant(indep)
    outcome = experimental_df[y].values

    model = sm.OLS(outcome, X, missing='drop')
    results = model.fit()
    coeff = round(results.params[1],3)
    std_error = round(results.bse[1],3)
    p_val = round(results.pvalues[1],3)
    N = results.nobs
    r2 = round(results.rsquared,3)
    adj_r2 = round(results.rsquared_adj,3)
    print()
    print(f"regression results for {y}")
    print()
    print(f"Delta (ATE): {coeff}")
    print(f"Std Error: {std_error}")
    print(f'p value: {p_val}')
    print(f"N: {N}")
    print(f"R2: {r2}")
    print(f"adj-R2: {adj_r2}")
    if p_val <= 0.05:
        print(f"{y} is significant at 95% Level")
    elif p_val <= 0.1:
        print(f"{y} is significant at 90% Level")
    else:
        print(f"{y} is not significant at 95% or 90 % Level")
    print()
    
```
